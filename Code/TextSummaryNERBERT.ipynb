# NAMED ENTITY RECOGNITION:

1. Entitas yang diberi nama adalah kategori yang telah ditentukan sebelumnya yang dipilih berdasarkan kasus penggunaan seperti nama orang, organisasi, tempat, kode, notasi waktu, nilai moneter, dll.

2.  NER bertujuan untuk menetapkan kelas untuk setiap token (biasanya satu kata) secara berurutan. Oleh karena itu, NER juga disebut sebagai klasifikasi token.

**Berikut Daftar Entitas Yang Ada Pada Dataset**





1.   O: Menunjukkan bahwa kata tersebut bukan bagian dari entitas apa pun.
2.   B-GEO: Menandakan awal entitas lokasi geografis (geographical entity).
3.   I-GEO: Menandakan entitas lokasi geografis yang berlanjut (geographical entity).
4.   B-GPE: Menandakan awal entitas negara, wilayah, atau tempat geografis politik (geopolitical entity).
5.   I-GPE: Menandakan entitas negara, wilayah, atau tempat geografis politik yang berlanjut (geopolitical entity).
6.   B-PER: Menandakan awal entitas nama orang (person).
7.   I-PER: Menandakan entitas nama orang yang berlanjut (person).
8.   B-ORG: Menandakan awal entitas organisasi (organization).
9.   I-ORG: Menandakan entitas organisasi yang berlanjut (organization).
10.  B-TIM: Menandakan awal entitas waktu (time).
11.  I-TIM: Menandakan entitas waktu yang berlanjut (time).
12.  B-NAT: Menandakan awal entitas asal-usul alam (natural).
13.  I-NAT: Menandakan entitas asal-usul alam yang berlanjut (natural).
14.  B-EVE: Menandakan awal entitas peristiwa (event).
15.  I-EVE: Menandakan entitas peristiwa yang berlanjut (event).
16.  B-Art: Menandakan awal entitas Karya seni (Artafact)
17.  I-Art: Menandakan entitas karya seni yang berlanjut (Artafact).

## Preprocessing


```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data["Sentence#"] = LabelEncoder().fit_transform(data["Sentence#"] )
```

```python
data
```

```python
data.rename(columns={"Sentence#":"sentence_id","Word ":"words","Tag":"labels"}, inplace =True)
```

```python
data["labels"] = data["labels"].str.upper()
```

```python
data
```

```python
X= data[["sentence_id","words"]]
Y =data["labels"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)
```

```python
#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})
```

```python
train_data
```

## Model Training

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
labels = data["labels"].unique().tolist()
labels
```

```python
args = NERArgs()
args.num_train_epochs = 1
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 16
args.eval_batch_size = 16
```

```python
model = NERModel('bert', 'bert-base-cased',labels=labels,args =args)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```

```python
print("Hasil Akurasi)",result)
```

```python
cleaned_text = ""
for token in datapred[0]:
    for word, label in token.items():
        if label != 'O':
            cleaned_text += f"{Fore.BLUE}{word}{Style.RESET_ALL} "
        else:
            cleaned_text += word + " "
    cleaned_text += " "
cleaned_text = cleaned_text.strip()
print("Hasil dari Ner :",cleaned_text)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

tokenizer = T5Tokenizer.from_pretrained("cahya/t5-base-indonesian-summarization-cased")
model = T5ForConditionalGeneration.from_pretrained("cahya/t5-base-indonesian-summarization-cased")
```

```python
def generate_summary(text):
    inputs = tokenizer.encode("ringkasan: " + text, truncation=True, max_length=512, return_tensors="pt")
    summary_ids = model.generate(inputs, num_beams=4, max_length=150, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
text_sum= teks
```

```python
summary = generate_summary(text_sum)
```
